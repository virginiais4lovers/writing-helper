# Writing Helper — Spec & Documentation

A collection of Claude Code skills that help people communicate more effectively using established communication frameworks. Built with two complementary modes:

- **Coaching skills** — Apply a framework to your writing right now
- **Practice skill** — Learn frameworks through deliberate practice grounded in learning science

The practice mode adapts evidence-based learning techniques from Dr. Cat Hicks' [Learning Opportunities](https://github.com/DrCatHicks/learning-opportunities) framework (CC-BY-4.0).

---
## 1. Problem Statement

### The writing gap

Most people know *what* they want to say but struggle with *how* to say it effectively. Communication frameworks like the Pyramid Principle or SCQA exist, but they live in books and workshops — disconnected from the moment someone is actually writing.

### The AI writing trap

AI tools can rewrite your text to be clearer, but this creates the same risks Dr. Cat Hicks identifies in AI-assisted coding:

| Risk | In coding | In writing |
|------|-----------|------------|
| **Generation effect undermined** | Accepting generated code without understanding it | Accepting rewritten text without knowing *why* it's better |
| **Fluency illusion** | Clean code feels understood | A polished revision feels like "yes, that's what I meant" |
| **Metacognition suppressed** | No time to notice skill gaps | No time to notice patterns (always burying the lead, mixing facts with opinions) |
| **Transfer blocked** | Can't apply patterns in new contexts | Can't write clearly without AI next time |

### Our approach

Two modes that address different needs:

1. **Coach mode** — When you have real writing to improve *now*, a coaching skill analyzes your draft through a framework's lens and suggests revisions. This is the "apply" mode.

2. **Practice mode** — When you want to *learn* the frameworks, interactive exercises use prediction, generation, and reflection to build genuine skill. This is the "learn" mode.

The insight: coaching alone creates dependency. Practice alone lacks immediate utility. Together, they form a learning loop.

## 3. Communication Frameworks

Five frameworks chosen to cover distinct communication needs:

### 3.1 Ladder of Inference (Chris Argyris)

**What it addresses:** Reasoning transparency — making the path from observation to conclusion visible.

**The model:**

<img src="images/ladder-of-inference.svg" width="300">

**Best for:** Analytical writing, persuasive arguments, feedback, any writing where you need readers to follow your reasoning.

**Coaching approach:**
1. Identify the core claim
2. Walk down the ladder — find where each rung is in the writing
3. Find inferential leaps (skipped rungs)
4. Suggest revisions that make reasoning explicit

**Common anti-pattern:** Jumping from selected data straight to conclusions, presenting interpretations as facts.

### 3.2 Pyramid Principle (Barbara Minto)

**What it addresses:** Logical structure — organizing ideas so the reader gets the point immediately.

**The model:**

<img src="images/pyramid-principle.svg" width="550">

**Three rules:**
1. Each level summarizes the level below (synthesis, not labeling)
2. Ideas in a group must be logically alike
3. Ideas in a group must be logically ordered (time, structure, or importance)

**MECE check:** Groups should be Mutually Exclusive (no overlap) and Collectively Exhaustive (no gaps).

**Best for:** Documents, presentations, proposals, status updates, any structured communication.

**Common anti-pattern:** Bottom-up writing where the conclusion comes last ("The Buildup").

### 3.3 SCQA (Situation-Complication-Question-Answer)

**What it addresses:** Narrative framing — creating a pull that makes readers want your answer.

**The model:**

<img src="images/scqa.svg" width="600">

**Key transitions:**
- S --> C should feel like "but..." or "however..."
- C --> Q should feel inevitable
- Q --> A is where the Pyramid Principle takes over

**Best for:** Proposals, problem statements, pitches, any writing that needs buy-in.

**Common anti-pattern:** Weak complication (no urgency), or jumping straight to the answer without establishing shared context.

### 3.4 BLUF (Bottom Line Up Front)

**What it addresses:** Directness — ensuring busy readers get the point immediately.

**The model:**

<img src="images/bluf.svg" width="350">

**Four anti-patterns the coaching skill diagnoses:**

| Pattern | Description | Common in |
|---------|-------------|----------|
| The Buildup | Background --> reasoning --> conclusion | Academic/cautious writers |
| The Narrative | Chronological story --> eventual point | Status updates |
| The Hedge | Over-qualified, unclear point | Sensitive topics |
| The Kitchen Sink | Multiple points, no hierarchy | Stream-of-consciousness |

**Best for:** Emails, Slack messages, executive communication, any writing for busy audiences.

**When NOT to use:** Bad news requiring empathy first, hostile audiences needing to be brought along, cultural contexts where directness is counterproductive.

### 3.5 Nonviolent Communication (Marshall Rosenberg)

**What it addresses:** Sensitive communication — expressing needs without blame or demand.

**The model:**

<img src="images/nvc.svg" width="600">

**Key distinctions:**

| Component | Not this | But this |
|-----------|----------|----------|
| Observation | "You're always late" | "You joined 10-15 min late to the last 3 standups" |
| Feeling | "You make me frustrated" | "I feel frustrated" |
| Need | "I need you to be on time" (strategy) | "I need predictability in our schedule" (need) |
| Request | "Be on time" (demand) | "Would you be willing to join by 9:05?" (request) |

**Best for:** Feedback, disagreements, difficult conversations, any writing with emotional stakes.

**Important nuance:** NVC is not about being "nice" — it's about being honest AND empathetic simultaneously. The need component is what makes it powerful; without it, it's just softened criticism.

## 5. Coaching Skills — Design Spec

### Interaction model

All five coaching skills follow the same pattern:

<img src="images/coaching-interaction.svg" width="400">

### Design principles for coaching skills

1. **Show, don't just tell** — Provide rewritten sentences, not abstract advice
2. **Use the writer's own words** — Reference specific phrases from their draft
3. **Name the pattern** — Give the problem a name they can recognize next time (e.g., "The Buildup")
4. **Collaborative tone** — Help them see their own reasoning, don't prove them wrong
5. **No draft? Interview.** — If invoked without arguments, guide them interactively
6. **Know when to stop** — If writing is already well-structured, say so and suggest refinements

## 6. Practice Skill — Design Spec

### The learning science foundation

The practice skill adapts techniques from [Learning Opportunities](https://github.com/DrCatHicks/learning-opportunities) by Dr. Cat Hicks. The core insight: **conditions that make learning slower in the short term often produce better long-term retention and transfer** (Bjork & Bjork, "desirable difficulties").

#### Risks the practice skill addresses

| Risk | What happens | How we counter it |
|------|-------------|-------------------|
| Generation effect undermined | User reads AI revision passively | User generates their own revision first |
| Fluency illusion | Polished output feels like understanding | Prediction exercises expose actual understanding |
| Metacognition suppressed | No awareness of recurring weaknesses | Reflection prompts surface patterns |
| Transfer blocked | Can only write well with AI help | Framework exercises build portable mental models |

### Core mechanism: Pause for input

The most important design decision: **end the message immediately after the question.** No hints, no examples, no encouragement that steers the answer.

<img src="images/pause-for-input.svg" width="400">

**Why this works:** Active retrieval strengthens memory traces in ways passive review does not (Roediger & Karpicke, 2006). The effort of generating an answer — even a wrong one — primes the mind to encode the correct information more effectively (Giebl et al., 2021).

**What's NOT allowed after the question:**
- Suggested responses or examples
- Hints disguised as encouragement ("Think about your audience...")
- Multiple questions stacked together
- Italicized clues

**What IS allowed:**
- Content-free reassurance: "(Your instinct is useful data here.)"
- Escape hatch: "(Or we can skip this one.)"

### Exercise types

Six exercise types, each adapted for writing from the learning science literature:

| Exercise type | Pattern | Learning mechanism |
|---------------|---------|-------------------|
| **Prediction → Observation → Reflection** | "What do you think the issue is?" → analyze together → "What surprised you?" | Pre-testing primes encoding (Giebl et al.) |
| **Generation → Comparison** | "Try rewriting this yourself" → show framework version → "What's different?" | Generation effect (active production > passive consumption) |
| **Diagnose this** | "What's wrong here and how would you fix it?" | Error analysis + retrieval practice |
| **Teach it back** | "Explain this framework to a colleague" | Self-explanation deepens understanding |
| **Audience shift** | "Rewrite this for a different audience" | Transfer through varied practice contexts |
| **Framework selection** | "Which framework would you use here and why?" | Metacognition + interleaving |

### Exercise examples by framework

Each framework has 3 exercises that progress from recognition to generation:

#### Ladder of Inference
1. **Spot the leap** — Find where a writer jumps rungs
2. **Separate the layers** — Sort sentences into observation / interpretation / conclusion
3. **Rewrite with transparency** — Make the reasoning visible

#### Pyramid Principle
1. **Find the buried lead** — Locate the main point in bottom-up writing
2. **Group the arguments** — Organize jumbled points into MECE groups
3. **Build a pyramid from scratch** — Create a full structure for a scenario

#### SCQA
1. **Identify the elements** — Label S, C, Q, A in a passage
2. **Strengthen the complication** — Rewrite a weak complication with urgency
3. **Reframe for audience** — Change the framing for a different reader

#### BLUF
1. **Find the bottom line** — Extract the point from a buried-lead message
2. **Diagnose the pattern** — Identify the anti-pattern (Buildup, Narrative, Hedge, Kitchen Sink)
3. **When NOT to BLUF** — Evaluate whether directness fits the context

#### NVC
1. **Observation vs. evaluation** — Sort statements into pure observations vs. judgments
2. **Find the need** — Identify the underlying need behind frustrated writing
3. **Transform the message** — Rewrite using all four NVC components

### Cross-framework exercises

These exercises build the meta-skill of knowing *when* to use *which* framework:

- **Framework selection** — Given a scenario, choose the right framework and justify
- **Framework combination** — For complex writing tasks, combine multiple frameworks
- **Audience shift** — Same content, different context: what framework shift is needed?

### Fading scaffolding

Guidance level adjusts based on demonstrated familiarity:

```
Scaffolding level    Example prompt
-----------------    --------------------------------------------------
High (early)         "The Pyramid Principle says to lead with your
                      main point. Looking at this draft, where is
                      the main point currently located?"

Medium (later)        "What would Minto say about this structure?"

Low (eventually)      "What's the issue here?"
```

**Fading adjusts the question setup, not the answer.** The learner always generates the answer themselves. If struggling, move back UP the scaffolding ladder — never hint at the answer.

### Session guardrails

| Rule | Rationale |
|------|-----------|
| Always ask before starting an exercise | Respect autonomy; learning requires buy-in |
| Max 2 exercises per session | Spacing effect: distributed practice > cramming |
| Stop if user declines once | Don't nag; one offer per session |
| Keep exercises to 10-15 minutes | Respect time; short focused practice > long sessions |
| End with transfer prompt | "When might you use this at work this week?" |

## 8. What Makes Writing Exercises Different

The practice skill adapts learning science from coding to writing, but writing has unique characteristics:

| Dimension | Coding exercises | Writing exercises |
|-----------|-----------------|-------------------|
| **Correctness** | Code compiles or doesn't; tests pass or fail | No compiler — quality depends on audience, context, purpose |
| **Valid approaches** | Usually one optimal solution | Multiple valid framings for the same message |
| **Emotional stakes** | Usually low (it's code) | Often high (real relationships, real consequences) |
| **Transfer goal** | Apply patterns to new codebases | Apply frameworks to any future communication |
| **Feedback style** | "This is incorrect because..." | "This could work, but consider..." (more nuanced) |

These differences shape exercise design:
- Exercises explore **tradeoffs** rather than enforcing single answers
- NVC exercises are handled with extra care (real emotional situations)
- Feedback acknowledges multiple valid approaches while still being direct about anti-patterns

## 9. Usage Examples

### Coaching: Apply BLUF to an email draft

```
User: /bluf
      Hi team, I wanted to give everyone an update on Project Aurora.
      As you know, we started the migration in January and the first
      phase went smoothly. However, in the second phase we hit some
      unexpected issues with the legacy API compatibility layer.
      The team has been working hard to resolve these, but we've
      determined that we need an additional two weeks to complete
      the migration safely. I'd like to discuss the revised timeline
      at Thursday's standup.

Coach: Your bottom line is buried in the second-to-last sentence.
       This follows "The Buildup" pattern — background → reasoning
       → conclusion.

       BLUF version:
       "Project Aurora's migration needs two more weeks — I'd like
       to discuss the revised timeline at Thursday's standup.

       Phase 2 hit unexpected issues with legacy API compatibility.
       The team is resolving them, but completing the migration
       safely requires extending our timeline from March 15 to
       March 29."
```

### Practice: Learn the Pyramid Principle

```
User: /writing-practice pyramid

Skill: Want to do a quick exercise on the Pyramid Principle?
       About 10-15 minutes.

User: Sure

Skill: Here's a passage from a project update:

       "We've been evaluating three vendors for the new analytics
       platform. Vendor A has the best pricing but limited
       customization. Vendor B offers strong customization but
       requires a 3-year commitment. Vendor C is the most expensive
       but includes migration support and has the strongest security
       certifications. Our compliance team flagged security as
       non-negotiable. The migration from our current platform is
       estimated at 6 months of engineering time without vendor
       support."

       > **Your turn:** What's the single main point of this
       > passage? Where did you find it?
       >
       > (Take your best guess.)

User: I think the main point is that they should go with Vendor C
       because of security and migration support, but it's never
       actually stated.

Skill: That's exactly right — and you've identified the core
       Pyramid Principle problem. The governing thought (recommend
       Vendor C) is implied but never stated. A reader has to
       assemble it themselves from the scattered evidence.

       > **Your turn:** Rewrite the opening sentence so it states
       > that main point directly.
       >
       > (Or we can skip this one.)
```

## 10. Future Directions

### Additional frameworks to consider

| Framework | What it adds | Priority |
|-----------|-------------|----------|
| **Toulmin Model** | Rigorous argumentation (claim, grounds, warrant, backing, qualifier, rebuttal) | Medium — overlaps with Ladder of Inference |
| **Monroe's Motivated Sequence** | Persuasive writing (attention, need, satisfaction, visualization, action) | Medium — overlaps with SCQA |
| **SBI** (Situation-Behavior-Impact) | Focused feedback structure | Low — largely covered by NVC |
| **PREP** (Point-Reason-Example-Point) | Simple persuasion | Low — largely covered by Pyramid |

### Potential enhancements

- **Plugin packaging** — Package as a Claude Code plugin (like learning-opportunities) for marketplace distribution
- **Orientation integration** — Pair with the `orient` plugin to generate writing-specific orientation for new projects ("how does this team communicate?")
- **Real writing triggers** — Auto-offer coaching after significant writing events (like the `learning-opportunities-auto` hook)
- **Web app companion** — Visual teaching tool for the frameworks (addresses the "no visual scaffolding" limitation of the skill format)
- **Progress tracking** — Lightweight memory of which frameworks a user has practiced and where they tend to struggle
